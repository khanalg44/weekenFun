## Spirographs

https://www.mathematische-basteleien.de/spirographs.htm

```python
def Epicycloids(r, R, a, t):
    factor = np.pi
    x = (R+r)*cos(r*t*factor/R) - a*cos((1+r/R)*t*factor)
    y = (R+r)*sin(r*t*factor/R) - a*sin((1+r/R)*t*factor)
    return x, y
```

```python
R = 5;
r = 1.005;
a = 2;

pts = np.array([Epicycloids(r, R, a, t) for t in np.linspace(0, 2*R, 400)])
fig, ax = plt.subplots(1,1, figsize=(7,7))
plt.plot(pts[:,0], pts[:,1], color="purple")
plt.xticks([]); plt.yticks([]);
```

```python
R = 10;
r = 1.005;
N = 30
aList = range(1, N+1)
fig, ax = plt.subplots(N//10, 10, figsize=(30, 3*N//10))
for ia, a in enumerate(aList):
    axi = ax[ia//10, ia%10]
    pts = np.array([Epicycloids(r, R, a, t) for t in np.linspace(0,2*R,400)])
    axi.plot(pts[:,0], pts[:,1], color="violet", lw=2.)
    axi.set_xticks([]); axi.set_yticks([]);
#axi.show()
```

```python
def main(r, R, N, Npts):
    aList = range(1, N+1)
    fig, ax = plt.subplots(N//10, 10, figsize=(30, 3*N//10))
    for ia, a in enumerate(aList):
        axi = ax[ia//10, ia%10]
        pts = np.array([Epicycloids(r, R, a, t) for t in np.linspace(0,1.8*R,Npts)])
        axi.plot(pts[:,0], pts[:,1], color="blue", lw=2.)
        axi.set_xticks([]); axi.set_yticks([]);
        
R = 10; r = 1.1; N = 30; Npts = 100
main(r, R, N, Npts)
```
