## Import File, arrange in Ascending Order based on Full Name

```python
import numpy as np
import pandas as pd
##!pip3 install xlrd
df = pd.read_excel('./data/2020-AbstractSubmissionForm.xlsx')
df = df.fillna('NAN')
df['Full Name'].sort_values()
df.head(10)
```

## Find Identical Entries and Remove them

```python
#Find rows with same First Name or Title
# Selects duplicate rows except first occurrence
dup_df = df[df.duplicated(['Full Name'])]
to_delete=[df[df.duplicated(['Full Name'])].index]
print (dup_df)
#to_delete,to_delete[0][0]
l=[]
for i in range(len(to_delete[0])):
    l.append(to_delete[0][i])
l
```

```python
#Drop duplicate rows based on row number
df=df.drop(df.index[l])
#Check Again for duplicates
df[df.duplicated(['Full Name'])]
```

## Assign Abstract Number 

```python
#Assign Abstract number in format ANPA2020_0001
ab_num =[]
for i in range(len(df)):
    ab_num.append("ANPA2020_0"+str(i+1).zfill(3))
    
ab_num[0:12]
```

```python
df['Abstract Number']=ab_num
print (df.columns)
df.to_csv('abstract_with_abs_numbers.csv')
#Keep only relevant columns
df=df[['Abstract Number','Full Name', 'Current Affiliation', 'Author List ', 'Author Affiliation/s',
       'Please choose field of your presentation', 'Abstract',
       'Title of the Presentation']]
df.head(3)
```

```python
#Adding names and affiliations
df['Authors']=df['Full Name']+' (1)'+ ' , '+df['Author List ']
df['Affiliations']=df['Current Affiliation']+' (1)' + ' , ' + df['Author Affiliation/s']
df.head(3)
```

```python
print (df.columns)
df_final=df[['Abstract Number', 'Please choose field of your presentation',
       'Abstract', 'Title of the Presentation', 'Authors', 'Affiliations']]
df_final.rename(columns = {'Please choose field of your presentation':'Field'}, inplace = True) 
df_final.head(2)
```
